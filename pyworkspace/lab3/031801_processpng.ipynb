```python
import cv2
import matplotlib.pyplot as plt
import numpy as np
img=cv2.imread('image.png')
img=cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
edges = cv2.Canny(img,100,200)
kernel = np.ones((3,3), np.uint8)

# Dilate the edges to make them wider
dilated_edges = cv2.dilate(edges, kernel, iterations=3)
dilated_edges=255-dilated_edges
# img=cv2.
plt.figure()
plt.imshow(img,cmap='gray')
plt.show()
plt.figure()
plt.imshow(dilated_edges,cmap='gray')
dilated_edges=cv2.cvtColor(dilated_edges,cv2.COLOR_GRAY2RGB)
plt.imsave('edge_cxk.png',dilated_edges)
plt.imshow(dilated_edges)
plt.show()
```
